```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make charts look beautiful
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("✅ Libraries loaded!")
```

```python
df = pd.read_csv('C:/Users/rishi_1rsw9tn/OneDrive/Documents/cricket-performance-predictor/data/cricket_players.csv')

print("Shape:", df.shape)
print("\nColumns:", list(df.columns))
print("\nFirst 5 rows:")
df.head()
```

```python
print("Basic Statistics:")
df.describe()
```

```python
plt.figure(figsize=(12, 6))
colors = sns.color_palette("husl", len(df))
bars = plt.bar(df['player'], df['runs'], color=colors)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.title('🏏 IPL Career Runs - Top Players', fontsize=16, fontweight='bold')
plt.xlabel('Player', fontsize=12)
plt.ylabel('Total Runs', fontsize=12)
plt.tight_layout()
plt.savefig('../data/top_scorers.png', dpi=150)
plt.show()
print("✅ Chart saved!")
```

```python
plt.figure(figsize=(10, 7))
scatter = plt.scatter(df['avg'], df['strike_rate'], 
                      s=df['runs']/30, 
                      c=df['sixes'], 
                      cmap='YlOrRd', 
                      alpha=0.8,
                      edgecolors='black',
                      linewidth=0.5)

# Add player name labels
for i, row in df.iterrows():
    plt.annotate(row['player'], 
                (row['avg'], row['strike_rate']),
                fontsize=7,
                xytext=(5, 5),
                textcoords='offset points')

plt.colorbar(scatter, label='Number of Sixes')
plt.title('⚡ Strike Rate vs Batting Average\n(bubble size = total runs)', 
          fontsize=14, fontweight='bold')
plt.xlabel('Batting Average', fontsize=12)
plt.ylabel('Strike Rate', fontsize=12)
plt.tight_layout()
plt.savefig('../data/strike_vs_avg.png', dpi=150)
plt.show()
print("✅ Chart saved!")
```

```python
plt.figure(figsize=(10, 8))
numeric_cols = ['matches', 'innings', 'runs', 'avg', 'strike_rate', 
                'hundreds', 'fifties', 'highest_score', 'fours', 'sixes']
correlation = df[numeric_cols].corr()

sns.heatmap(correlation, 
            annot=True, 
            fmt='.2f',
            cmap='coolwarm',
            center=0,
            square=True,
            linewidths=0.5)

plt.title('🔥 Correlation Heatmap of Player Stats', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../data/correlation_heatmap.png', dpi=150)
plt.show()
print("✅ Heatmap saved!")
```

```python
x = range(len(df['player']))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar([i - width/2 for i in x], df['fours'], width, 
               label='Fours', color='steelblue', alpha=0.8)
bars2 = ax.bar([i + width/2 for i in x], df['sixes'], width, 
               label='Sixes', color='coral', alpha=0.8)

ax.set_xlabel('Player', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.set_title('🏏 Fours vs Sixes per Player', fontsize=14, fontweight='bold')
ax.set_xticks(list(x))
ax.set_xticklabels(df['player'], rotation=45, ha='right', fontsize=9)
ax.legend(fontsize=11)
plt.tight_layout()
plt.savefig('../data/fours_vs_sixes.png', dpi=150)
plt.show()
print("✅ Chart saved!")
```
